```python
from folium import plugins
import folium
import pandas as pd
import branca 
# from streamlit_folium import st_folium


def make_marker_and_add_to_featuregroup(featuregroup_name, popup_data, location, name):
    iframe = branca.element.IFrame(html=popup_data , width=300, height=400)
    popup = folium.Popup(iframe, max_width=300)
    marker = folium.Marker(
                location=location, popup=popup, tooltip=name)\
                .add_to(feature_group[featuregroup_name])

# reading data
df = pd.read_excel("Sample_feed.xlsx")
samples = df.to_dict('records')

# making the map 
map = folium.Map(location=[50.141327, 8.692665],zoom_start=13)

# making a feature group for each unique feature (saving to a dict)
features_list = ['Water', 'Gas', 'Air ventilator', 'Bike storage',
                 'Electricity', 'Charging possibility for vehicles', 
                 'Loading/unloading access', 'Fridge', 'Table', 'Extra Rooms',
                 'Parking', 'Chair', 'Shelves', 'Forklifts',
                 'Green Space<=200', 'Green Space<=400', 'Green Space<=600',
                 'Covered Space<=200', 'Covered Space<=400', 'Covered Space<=600',
                 'Rent<=1500', 'Rent<=2500', 'Rent<=3500',
                 'Height<=2', 'Height<=4', 'Height<=6',
                 'Cost<=30', 'Cost<=60', 'Cost<=90',]

feature_group = {}
for feature in features_list:
  if feature == 'Forklifts':
    feature_group[feature] = folium.FeatureGroup(name=f'<span>{feature}</span><br><br><i style="color:purple;">Green Space</i><br>', show=False)
  elif feature == 'Green Space<=200':
    feature_group[feature] = folium.FeatureGroup(name=f'<=200', show=True,)
  elif feature == 'Green Space<=400':
    feature_group[feature] = folium.FeatureGroup(name=f'<=400', show=False)
  elif feature == 'Green Space<=600':
    feature_group[feature] = folium.FeatureGroup(name=f'<span><=600</span><br><br><i style="color:purple;">Covered Space</i><br>', show=False)
  elif feature == 'Covered Space<=200':
    feature_group[feature] = folium.FeatureGroup(name=f'<=200 ', show=True)
  elif feature == 'Covered Space<=400':
    feature_group[feature] = folium.FeatureGroup(name=f'<=400 ', show=False)
  elif feature == 'Covered Space<=600':
    feature_group[feature] = folium.FeatureGroup(name=f'<span><=600</span><br><br><i style="color:purple;">Rent</i><br>', show=False)
  elif feature == 'Rent<=1500':
    feature_group[feature] = folium.FeatureGroup(name=f'<=1500', show=True)
  elif feature == 'Rent<=2500':
    feature_group[feature] = folium.FeatureGroup(name=f'<=2500', show=False)
  elif feature == 'Rent<=3500':
    feature_group[feature] = folium.FeatureGroup(name=f'<span><=3500</span><br><br><i style="color:purple;">Height</i><br>', show=False)
  elif feature == 'Height<=2':
    feature_group[feature] = folium.FeatureGroup(name=f'<=2', show=True)
  elif feature == 'Height<=4':
    feature_group[feature] = folium.FeatureGroup(name=f'<=4', show=False)
  elif feature == 'Height<=6':
    feature_group[feature] = folium.FeatureGroup(name=f'<span><=6</span><br><br><i style="color:purple;">Cost/sqm</i><br>', show=False)
  elif feature == 'Cost<=30':
    feature_group[feature] = folium.FeatureGroup(name=f'<=30', show=True)
  elif feature == 'Cost<=60':
    feature_group[feature] = folium.FeatureGroup(name=f'<=60', show=False)  
  elif feature == 'Cost<=90':
    feature_group[feature] = folium.FeatureGroup(name=f'<=90', show=False)
  else:
    feature_group[feature] = folium.FeatureGroup(name=f'<span>{feature}</span>', show=False).add_to(map)


# reading all features
for s in samples:
  image1 = f'{s["Image1"]}'
  image2 = f'{s["Image2"]}'
  image3 = f'{s["Image3"]}'
  image4 = f'{s["Image4"]}'
  name = f'{s["Name"]}'
  address = f'{s["Address"]}'
  email = f'{s["Email"]}'
  phone = f'{s["Phone"]}'
  electricity = f'{s["Electricity"]}'
  gas = f'{s["Gas"]}'
  water = f'{s["Water"]}'
  bike_storage = f'{s["Bike storage"]}'
  air_ventilator = f'{s["Air Ventilator"]}'
  cost = f'{s["Cost/sqm"]}'
  lan, lon = s["Lat"], s["Lon"]
  green_space = f'{s["Space(green area)"]}'
  covered_space = f'{s["Space(covered area)"]}'
  height = f'{s["Height"]}'
  rent = f'{s["Total rent/month"]}'
  charging = f'{s["Charging possibility for vehicles"]}'
  loading_access = f'{s["Loading/unloading access"]}'
  shelves = f'{s["Shelves"]}'
  forklifts = f'{s["Forklifts"]}'
  fridge = f'{s["Fridge"]}'
  table = f'{s["Table"]}'
  chair = f'{s["Chair"]}'
  parking_space = f'{s["Parking spaces"]}'
  extra_rooms = f'{s["Extra rooms"]}'


  popup_data = f"""
              <html>

                <head>
                    <script src='https://kit.fontawesome.com/e793be24ac.js' crossorigin='anonymous'></script>
                    <!-- Latest compiled and minified CSS -->
                    <link rel='stylesheet' href='https://maxcdn.bootstrapcdn.com/bootstrap/3.4.1/css/bootstrap.min.css'>

                    <!-- jQuery library -->
                    <script src='https://ajax.googleapis.com/ajax/libs/jquery/3.6.0/jquery.min.js'></script>

                    <!-- Latest compiled JavaScript -->
                    <script src='https://maxcdn.bootstrapcdn.com/bootstrap/3.4.1/js/bootstrap.min.js'></script>
                </head>

                <body>
                    <div class='row'>
                        <div class='col-sm-4'>
                            <div class='panel panel-default'>
                                <div class='panel-heading'>
                                    <h3>{name}</h3>
                                </div>
                                <div class='panel-body'>
                                    <span font-size:4=''>
                                        <img style=' display: block;
                                        margin-left: auto;
                                        margin-right: auto;' src={image1} width='200' height='200'>
                                        <br>
                                        <br>
                                        <i class='fa-solid fa-envelope'></i> <b> E-mail: </b> <a href='mailto:{email}'>{email}</a> <br>
                                        <i class='fa-solid fa-phone'></i> <b> Phone: </b> <a href='tel:{phone}'>{phone}</a> <br>
                                        <i class='fa-solid fa-location-dot'></i> <b> Address: </b> {address} <br> <br>
                                    </span>
                                </div>
                            </div>
                        </div>
                    </div>
                    <div class='row'>
                        <div class='col-sm-4'>
                            <div class='panel panel-default'>
                                <div class='panel-heading'>
                                    <h2>Facility features</h2>
                                </div>
                                <div class='panel-body'>
                                    {"<i class='fa-xl fa-solid fa-bolt' aria-hidden='true' title='Electricity'></i>&nbsp;" if electricity=='Yes' else ' '}
                                    {"<i class='fa-xl fa-solid fa-fire-flame-simple' aria-hidden='true' title='Gas'></i>&nbsp;" if gas=='Yes' else ' '}
                                    {"<i class='fa-xl fa-solid fa-droplet' aria-hidden='true' title='Water'></i>&nbsp;" if water=='Yes' else ' '}
                                    {"<i class='fa-xl fa-solid fa-bicycle' aria-hidden='true' title='Bike Storage'></i>&nbsp;" if bike_storage=='Yes' else ' '}
                                    {"<i class='fa-xl fa-solid fa-fan' aria-hidden='true' title='Air Ventilator'></i>&nbsp;" if air_ventilator=='Yes' else ' '}
                                    {"<i class='fa-xl fa-solid fa-charging-station' aria-hidden='true' title='Charging'></i>&nbsp;" if charging=='Yes' else ' '}
                                    {"<i class='fa-xl fa-solid fa-truck-ramp-box' aria-hidden='true' title='Loading Access'></i>&nbsp;" if loading_access=='Yes' else ' '}
                                </div>
                            </div>
                        </div>
                    </div>
                    <div class='row'>
                        <div class='col-sm-4'>
                            <div class='panel panel-default'>
                                <div class='panel-heading'>
                                    <h2>Facility properties</h2>
                                </div>
                                <div class='panel-body'>

                                    <table class='table table-bordered'>
                                        <thead>
                                            <tr>
                                                <th>properties</th>
                                                <th>Availability</th>
                                            </tr>
                                        </thead>
                                        <tbody>
                                            <tr>
                                                <td>Shelves</td>
                                                <td>{shelves}</td>
                                            </tr>
                                            <tr>
                                                <td>Forklifts</td>
                                                <td>{forklifts}</td>
                                            </tr>
                                            <tr>
                                                <td>Fridge</td>
                                                <td>{fridge}</td>
                                            </tr>
                                            <tr>
                                                <td>Table</td>
                                                <td>{table}</td>
                                            </tr>
                                            <tr>
                                                <td>Chair</td>
                                                <td>{chair}</td>
                                            </tr>
                                            <tr>
                                                <td>Parking space</td>
                                                <td>{parking_space}</td>
                                            </tr>
                                            <tr>
                                                <td>Extra rooms</td>
                                                <td>{extra_rooms}</td>
                                            </tr>
                                            <tr>
                                                <td>Green Space</td>
                                                <td>{green_space} </td>
                                            </tr>
                                            <tr>
                                                <td>Covered Space</td>
                                                <td>{covered_space}</td>
                                            </tr>
                                            <tr>
                                                <td>Height</td>
                                                <td>{height}</td>
                                            </tr>
                                            <tr>
                                                <td>Rent</td>
                                                <td>{rent}</td>
                                            </tr>
                                            <tr>
                                                <td>Cost/sqm</td>
                                                <td>{cost}</td>
                                            </tr>
                                        </tbody>
                                    </table>
                                </div>
                            </div>
                        </div>
                    </div>
                </body>

              </html>

    """
  
  # making markers and adding to the corresponding feature group
  if 'Yes' in loading_access:
    make_marker_and_add_to_featuregroup('Loading/unloading access',
                                        popup_data, [lan, lon], name)

  if 'Yes' in charging:
    make_marker_and_add_to_featuregroup('Charging possibility for vehicles',
                                        popup_data, [lan, lon], name)

  if 'Yes' in water:
    make_marker_and_add_to_featuregroup('Water',
                                        popup_data, [lan, lon], name)
    
  if 'Yes' in gas:
    make_marker_and_add_to_featuregroup('Gas',
                                        popup_data, [lan, lon], name)

  if 'Yes' in electricity:
    make_marker_and_add_to_featuregroup('Electricity',
                                        popup_data, [lan, lon], name)
    
  if 'Yes' in air_ventilator:
    make_marker_and_add_to_featuregroup('Air ventilator',
                                        popup_data, [lan, lon], name)               

  if 'Yes' in bike_storage:               
    make_marker_and_add_to_featuregroup('Bike storage',
                                        popup_data, [lan, lon], name)
    
  if int(table) != 0:
    make_marker_and_add_to_featuregroup('Table',
                                        popup_data, [lan, lon], name)  
    
  if int(chair) != 0:
    make_marker_and_add_to_featuregroup('Chair',
                                        popup_data, [lan, lon], name)                

  if int(fridge) != 0:
    make_marker_and_add_to_featuregroup('Fridge',
                                        popup_data, [lan, lon], name)
    
  if int(extra_rooms) != 0:
    make_marker_and_add_to_featuregroup('Extra Rooms',
                                        popup_data, [lan, lon], name)
    
  if int(parking_space) != 0:
    make_marker_and_add_to_featuregroup('Parking',
                                        popup_data, [lan, lon], name)
  if int(forklifts) != 0:
    make_marker_and_add_to_featuregroup('Forklifts',
                                        popup_data, [lan, lon], name)
    
  if int(shelves[0]) != 0:
    make_marker_and_add_to_featuregroup('Shelves',
                                        popup_data, [lan, lon], name)                

  if float(green_space) <= 200:
    make_marker_and_add_to_featuregroup('Green Space<=200',
                                        popup_data, [lan, lon], name)
    
  if float(green_space) <= 400:
    make_marker_and_add_to_featuregroup('Green Space<=400',
                                        popup_data, [lan, lon], name)
    
  if float(green_space) <= 600:
    make_marker_and_add_to_featuregroup('Green Space<=600',
                                        popup_data, [lan, lon], name)  
    
  if float(covered_space) <= 200:
    make_marker_and_add_to_featuregroup('Covered Space<=200',
                                        popup_data, [lan, lon], name)
    
  if float(covered_space) <= 400:
    make_marker_and_add_to_featuregroup('Covered Space<=400',
                                        popup_data, [lan, lon], name)
    
  if float(covered_space) <= 600:
    make_marker_and_add_to_featuregroup('Covered Space<=600',
                                        popup_data, [lan, lon], name)
    
  if float(height) <= 2:
    make_marker_and_add_to_featuregroup('Height<=2',
                                        popup_data, [lan, lon], name)
    
  if float(height) <= 4:
    make_marker_and_add_to_featuregroup('Height<=4',
                                        popup_data, [lan, lon], name)  
      
  if float(height) <= 6:
    make_marker_and_add_to_featuregroup('Height<=6',
                                        popup_data, [lan, lon], name)
    
  if float(rent) <= 1500:
    make_marker_and_add_to_featuregroup('Rent<=1500',
                                        popup_data, [lan, lon], name)
    
  if float(rent) <= 2500:
    make_marker_and_add_to_featuregroup('Rent<=2500',
                                        popup_data, [lan, lon], name)
    
  if float(rent) <= 3500:
    make_marker_and_add_to_featuregroup('Rent<=3500',
                                        popup_data, [lan, lon], name)
    
  if float(cost) <= 30:
    make_marker_and_add_to_featuregroup('Cost<=30',
                                        popup_data, [lan, lon], name)
    
  if float(cost) <= 60:
    make_marker_and_add_to_featuregroup('Cost<=60',
                                        popup_data, [lan, lon], name) 
    
  if float(cost) <= 90:
    make_marker_and_add_to_featuregroup('Cost<=90',
                                        popup_data, [lan, lon], name)      
                                        
# add all feature groups to the map
for f in feature_group:
  map.add_child(feature_group[f])

# add extra features to the map
# folium.features.LatLngPopup().add_to(map)
plugins.LocateControl().add_to(map)
plugins.Geocoder().add_to(map)
# folium.TileLayer(tiles='Stamen Terrain').add_to(map)
# folium.TileLayer(tiles='cartodbpositron').add_to(map)
fmtr = "function(num) {return L.Util.formatNum(num, 3) + ' ยบ ';};"
plugins.MousePosition(separator=' | ', prefix="Mouse:",
              lat_formatter=fmtr, lng_formatter=fmtr).add_to(map)



popup = "<input type='checkbox' id='electricity'>electricity</input><br><input type='checkbox' id='water'>water</input><br><label for='forklifts'>forklifts</label><input type='text' id='myInput' placeholder='forklifts'></input><br><br><button onclick='myFunction()'>save location</button>"
print(type(popup))
# m = folium.features.ClickForMarker(popup=popup).add_to(map)
# print(m.get_bounds())
import jinja2

el = folium.MacroElement().add_to(map)
el._template = jinja2.Template("""
    {% macro script(this, kwargs) %}
    function myFunction() {
      var copyText = document.getElementById("myInput");

      /* Select the text field */
      copyText.select();
      copyText.setSelectionRange(0, 99999); /* For mobile devices */

      /* Copy the text inside the text field */
      document.execCommand("copy");
    }
    {% endmacro %}
""")

# folium.LayerControl().add_to(map)

map
map.save('map.html')
```
